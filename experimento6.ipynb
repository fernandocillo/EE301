Serão utilizados um espectrômetro, um prisma óptico, uma rede de difração, uma fonte luminosa composta por uma fonte de alta tensão e diversas lâmpadas espectrais.

Um espectrômetro óptico de emissão compõe-se essencialmente de um sistema colimador, um elemento de dispersão de luz e de um sistema analisador, como esquematizado na figura a seguir.  Sua função é analisar a luz proveniente de uma fonte luminosa.

download.png

- *Sistema colimador:* Geralmente é constituído por uma fenda de entrada de luz, cuja largura é ajustável, e de um conjunto de lentes para colimar a luz proveniente da fenda.

- *Elemento de dispersão:* Pode ser um prisma óptico ou uma rede de difração, usualmente do tipo de reflexão.

- *Sistema analisador:* Composto por um conjunto de lentes que forma imagens da fenda sobre a superfície de análise.  A análise pode ser feita visualmente ou com um sensor óptico (CCD, válvula fotomultiplicadora, etc).

Um sistema mecânico permite o deslocamento relativo entre o elemento de dispersão e o sistema analisador, possibilitando a observação e a medida das posições angulares e intensidades das diversas imagens da fenda formadas.

Para cada comprimento de onda visível presente na luz emitida pela fonte forma-se uma correspondente imagem.  Cada uma dessas imagens é denominada *linha espectral*.  Sua posição angular é função de seu comprimento de onda.  Para uma dada fonte luminosa, o conjunto de suas linhas espectrais forma seu *espectro luminoso*.  Certas fontes de luz, como as lâmpadas incandescentes, emitem, dentro de uma faixa, todos os comprimentos de onda; seu espectro é por isso chamado de *espectro contínuo*.  O espectro de emissão de gases, por possuir linhas espectrais definidas, é chamado *espectro discreto*.

### Especrometria de Rede de Difração

Substitua o prisma óptico por uma rede de difração do tipo transmissão, cuja superfície deverá ficar perpendicular ao feixe de luz proveniente do colimador.

**Atenção:** nunca toque na superfície da rede, pois isso a danificaria permanentemente.

Determine os ângulos de desvio das 4 linhas mais intensas do espectro visível da lâmpada de vapor de mercúrio.  Observe o espectro das outras lâmpadas disponíveis no laboratório.  Anotar as diferenças encontradas.

Foi possível observar mais ordens para a lâmpada de nitrogênio e menos ordem para a outra lâmpada .

Sabendo-se que o comprimento de onda da linha verde é de 546,1 nm, determine a constante de rede da rede de difração.

```python
# CÁLCULOS
```

$$
m⋅λ=d⋅sinθ\\
d=\frac{m⋅λ}{sinθ}\\
d=\frac{1\cdot546,1\cdot10^{-9}}{sin(19°20')}\\
d=1,65 \cdot 10^{-6}m
$$

```python
# DADOS COLETADOS
```

$$
\begin{aligned}
& \text {Tabela 2. Medidas de 2*θ}\\
&\begin{array}{|c|ccc|}
\hline \hline \text { Cor } & \text { Ângulo medido } & \text { 2*θ} & \text{ θ } \\
\hline Amarelo_1 & 181°30'-140° & 41°30'& 20°45'\\
Amarelo_2 & 181°40'-140°10' & 41°30' & 20°45'\\\
Verde & 341°-302°20' & 38°40 & 19°20'\\
Violeta & 176°20'-145°20' & 31° & 15°30'\\
\hline
\end{array}
\end{aligned}
$$

Com o valor da constante de rede calculado, determine o comprimento de onda das linhas medidas.

```python
# CÁLCULOS
```

$$ λ=\frac{d\cdot sin(θ)}{m} \\
Amarelo_1\\
λ=\frac{1,65\cdot 10^{-6}m\cdot sin(20°45')}{1}=5,844 \cdot 10^{-7}m = 584,4 nm\\
Amarelo_2\\
λ=\frac{1,65\cdot 10^{-6}m\cdot sin(20°45')}{1}=5,844 \cdot 10^{-7}m = 584,4 nm \\
 Violeta\\

$$

$$ Verde $$

$$ λ=\frac{1,65\cdot 10^{-6}m\cdot sin(19°20')}{1}=5,463\cdot 10^{-7}m=546,8nm $$



Usando os valores dos comprimentos de onda obtidos com o espectrômetro de rede, construa a curva de calibração do espectrômetro de prisma, ou seja, desvio angular versus comprimento de onda.

```python
# GRÁFICO

import matplotlib.pyplot as plt

x = [20.75,19.33,15.5]   #graus
y = [584.4,440.8,546.8]  #nm

plt.figure()
plt.plot(x,y,'o')
plt.xlabel('θ  [°]')
plt.ylabel('λ  [nm]')
plt.grid(ls='--')
plt.show()
```
