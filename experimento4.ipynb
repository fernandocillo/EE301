`PREENCHER` Os parâmetros geométricos que caracterizam os fenômenos observados são:
- A distância da fenda para o anteparo $L$;
- O comprimento de onda da luz indicente $\lambda$;
- A largura da fenda $a$ ou a constante de difração da rede utilizada $\Lambda$;
- O comprimento de um dos ou de mais de um padrão observado no anteparo $D$ e a ordem do(s) padrão(ões) utilizado(s) $m$; 
- O ângulo entre a normal do raio de luz e o padrão no anteparo $\theta$.

Tomando como padrão a réplica da rede de difração com 300 linhas/mm, determine o comprimento de onda da luz do laser.  Compare com o valor estabelecido de 632.8 nm.

```python
# CÁLCULOS
# parâmetros para o sistema
m = 1 # ordem
cte_rede = 300e3 # linhas/m
lambda_rede = 1/cte_rede
D = 56.4e-2 # m
L = 147.1e-2 # m

# cálculo do d e do ângulo
d = D/2
tan_theta = d/L
theta = np.arctan(tan_theta)

# comprimento de onda
lambda_luz = np.sin(theta)*lambda_rede/m
print('O comprimento de onda obtido para a luz do laser é de: {:.3e}'.format(lambda_luz))
```

`PREENCHER` Desta forma, o comprimento de onda para a luz do laser obtido com o arranjo experimental é de 627.6 nm.

Usando o valor estabelecido, determine as larguras das fendas simples e a constante da rede de difração do diapositivo.

```python
# CÁLCULOS

# rede de difração
m = 8 # ordem
lambda_luz = 632.8e-9
D = 17.9e-2 # m
L = 147.7e-2 # m

# cálculo do d e do ângulo
d = D/2
tan_theta = d/L
theta = np.arctan(tan_theta)

# cálculo da constante da rede
lambda_rede = (1/np.sin(theta))*(lambda_luz*m)
cte_rede = (1/lambda_rede)/1e3
print('A constante da rede vale: {:.5}'.format(cte_rede))

# fenda simples
m = 2
D = 10.0e-2
L = 146.5e-2

# cálculo da largura
d = D/2
tan_theta = d/L
theta = np.arctan(tan_theta)
a = (1/np.sin(theta))*lambda_luz*m

print('A largura da fenda simples é de: {:.3e}'.format(a))
```

Como pode ser observado, a constante da rede de difração é de 11.948 e a largura da fenda simples vale 37.10 $\mu$m.

Determine o diâmetro de um fio de cabelo.

```python
# CÁLCULOS
# parâmetros
m = 8
D = 16.0e-2
L = 142.0e-2
lambda_luz = 632.8e-9

# cálculos
d = D/2
tan_theta = d/L
theta = np.arctan(tan_theta)

# diâmetro
diametro = (1/np.sin(theta))*lambda_luz*m
print('O fio de cabelo utilizado tem a largura de: {:.3e}'.format(diametro))
```

O diâmetro obtido é de cerca de 90 $\mu$m.
