O modelo de onda clássico da luz, da teoria eletromagnética, prevê que a emissão deveria ocorrer independentemente do comprimento de onda, pela absorção contínua da radiação eletromagnética incidente nos elétrons, já que a onda transporta uma energia que é diretamente proporcional à sua intensidade, qualquer que seja a frequência.  Com o aumento da intensidade da luz aumentaria a energia dos elétrons emitidos e portanto o potencial de freamento.  Isto contradiz os dados experimentais.  Uma outra contradição com o modelo clássico é a inexistência de um tempo entre o instante em que o catodo é iluminado e a ocorrência da emissão.  Segundo o modelo clássico, a emissão deveria ocorrer após algum tempo de exposição para luz de baixa intensidade.

O efeito fotoelétrico só foi entendido no início do século XX, com o surgimento da ideia da quantização da energia.

Em 1901, Planck havia postulado a quantização da energia irradiada e absorvida por um sistema físico, isto é, o ganho ou a perda de energia de um sistema se daria pela absorção ou emissão de uma unidade ou quantum de energia irradiante, de magnitiude $E$ proporcional à frequência $f$ da radiação envolvida, expressa por $E = hf$.  O coeficiente de proporcionalidade $h$ tornou-se conhecido posteriormente como a constante de Planck.  No entanto, Planck considerava que a energia da própria radiação era contínua.  Apenas a absorção ou a emissão se daria de modo quântico.

Em trabalho publicado em 1905 sobre irradiação de cavidades, Einstein propôs que a luz se comportava como se fosse composta por uma coleção de quanta de energia discretos de magnitude $E = Cf$.  Ele chegou a esta conclusão a partir da termodinâmica estatística, utilizando a lei da radiação de Wien, que é precisa apenas nas altas frequências.  Considerou que a teoria de Maxwell era bastante eficaz na descrição dos fenômenos como interferência, refração e difração, mas percebeu que as observações ópticas se referiam a valores médios ao longo do tempo e não a valores instantâneos.  Deve-se salientar que apenas no ano seguinte Einstein percebeu a conexão da sua proposta com a teoria de Planck, e que $C = h$.  Em 1926, G. N. Lewis deu o nome de fóton a este quanta de energia de luz.

No mesmo ano de 1905, Einstein aplicou a ideia do quanta de luz no modelamento do efeito fotoelétrico.  No processo, a energia de um fóton é absorvida por um elétron no catodo.  Para que ocorra a emissão deste elétron é necessária uma energia mínima $W$ do fóton, denominada função de trabalho do particular metal que constitui o catodo.  É a energia mínima necessária para se extrair um elétron da superfície do catodo, uma característica do material.  Valores típicos da função de trabalho são de alguns elétrons-volts (eV).

Se a energia do fóton for maior ou igual a $W$, o elétron é extraído do catodo, gerando a fotocorrente.  Se a energia do fóton for maior que $W$, o excedente de energia vai corresponder à energia mecânica cinética do elétron, após a extração:

\begin{equation}
\frac{mv^2}{2} = hf - W
\end{equation}

O elétron, com massa $m$, sai do eletrodo com uma velocidade $v$.

Aplicando-se a voltagem constante negativa $V_F$, freia-se o elétron e, portanto, anula-se a fotocorrente.

Quando se aumenta a intensidade da luz monocromática, mais fótons incidem sobre o eletrodo e mais elétrons são emitidos, aumentando a corrente de saturação para voltagens positivas aplicadas pela fonte.  Mas a energia cinética dos elétrons extraídos não se altera e, portanto, a tensão de freamento $V_F$ também não se altera.

A energia potencial $eV_F$ do elétron, com carga $e$, corresponde à energia cinética dos elétrons extraídos:

\begin{equation}
eV_F = hf - W \Leftrightarrow V_F = \frac{hf}{e} - \frac{W}{e}
\end{equation}

O potencial de freamento $V_F$ é uma função linear da frequência dos fótons, o que concorda com os dados experimentais.  As medidas de $V_F$ em função das frequências da luz incidente permitem a determinação de $h$ e $W$.

### Notas Históricas

As primeiras observações experimentais documentadas relativas ao efeito fotoelétrico foram feitas, de modo não intencional, por Heinrich Herz em 1887.  Seu experimento, que teve pleno sucesso, tinha como objetivo demonstrar a existência das ondas eletromagnéticas propostas pela teoria eletromagnética de Maxwell em 1867.  Para gerar as ondas eletromagnéticas, Herz provocava faíscas entre dois eletrodos esféricos por uma descarga elétrica oscilante, produzida por um circuito LC.  As ondas eram então detetadas em uma antena circular, posicionada no foco de uma superfície metálica côncava, com dois eletrodos separados de modo a permitir a ressonância com a radiação emitida.  Na recepção da radiação ocorria uma faísca entre os eletrodos.  Ele observou que a faísca na detecção era mais vigorosa quando os eletrodos do detetor ficavam expostos à luz proveniente das faíscas que ocorriam no transmissor, predominantemente composta por comprimentos de onda nas regiões violeta e ultravioleta.  Posteriormente, em 1899, J. J. Thompson mostrou que a luz ejetava elétrons da superfície metálica dos eletrodos, e o fenômeno ficou conhecido então como efeito fotoelétrico.  No experimento de Herz, os elétrons ejetados, acelerados pelo campo elétrico entre os eletrodos, contribuíam na ionização do ar, facilitando a descarga.

É interessante notar que ao mesmo tempo em que Herz comprovava a teoria de Maxwell, pois demonstrou que a radiação produzida exibia todas as propriedades das ondas como reflexão, onda estacionária, interferência e refração, também descobria o que depois veio a ser o efeito fotoelétrico, que se tornou uma das primeiras comprovações experimentais do comportamento quântico da natureza.

Em 1902, Phillip Lenard, um discípulo de Herz, caracterizou o efeito fotoelétrico, utilizando uma placa emissora iluminada pela luz proveniente de uma descarga em arco entre eletrodos de carbono, de alta intensidade.  Observou, surpreso, tanto o efeito de freamento das voltagens negativas como a independência da energia dos elétrons emitidos da intensidade da luz.  Também observou que a energia dos elétrons dependia da cor, ou frequência da luz, características que contradiziam o que era esperado da teoria eletromagnética.

Nos experimentos iniciais para o estudo do efeito fotoelétrico foram utilizados catodos de metais alcalinos como Na e K, que apresentam maior eficiência de emissão.  Nestes materiais o efeito fotoelétrico já ocorre com luz visível.  A maioria dos metais necessitam ser iluminados com luz ultravioleta.  No entanto, estes materiais apresentam uma eficiência de emissão de apenas 0,1%.  Os eletrodos das células atuais possuem camadas superficiais de materiais como antimoneto de césio (CsSb), chegando à uma eficiência em torno de 20%.

### Função de trabalho do catodo $W$ e constante de Planck $h$

Refaça a calibração inicial para escala de corrente na posição 10⁻¹³.

1. Descubra a janela do fotodiodo e insira a capa com abertura de 4 mm de diâmetro, e o filtro de 365 mm.
1. Descubra a janela da fonte de luz de mercúrio.
1. Ajuste a voltagem pelo botão *voltage adjust* até que a fotocorrente seja zero e anote o valor indicado.
1. Cubra a fonte de luz de mercúrio.
1. Repita o procedimento de medida descrito com os filtros de 436 mm, 546 mm e 577 mm.
1. Trace o gráfico do potencial de freamento $V_F$ versus frequência da luz $f$.
1. Extraia do gráfico os valores da função de trabalho do catodo $W$ e da constante de Planck $h$.
1. Analise os resultados.

```python
# DADOS E ANÁLISE
```

```python
# Velocidade da luz no vácuo
c=299792458 #m/s

λ=[365, 436, 546, 577] # comprimentos de onda (em nm)

f=[0,0,0,0] # criando a lista das frequências (em Hertz)

# c=λ*f -> f = c/λ
for i in range(0, len(λ)):
  f[i]=10**9*c/λ[i]

# Potencial de freamento = 
# tensão máxima para que não haja passagem de corrente (mínima em módulo)
Vf=[-2.18, -1.59, -1.08, -.91]

# Gráfico
plt.plot(f, Vf, color='green', linewidth=2)
plt.plot(f, Vf, 'o', color='green')
plt.ylabel('Potencial de freamento (V)', fontsize=14)
plt.xlabel('Frequência da luz (Hz)',fontsize=14) 
plt.title('Curva de Vf x f', fontsize=20) 
plt.show()
```

Partindo da equação $eV_F = hf - W$, podemos obter os valores de $h$ e $W$ fazendo uma regressão linear para uma equação de primeiro grau.

Ou seja, $y=a*x+b$, em que:

*   y = e*Vf
*   x = f
*  a = h
* b = -W



```python
### REGRESSÃO LINEAR ###

#criando a variável e*Vf (y da equação)
e= -1.60217663*10**(-19)
eVf=[0,0,0,0]
for i in range(0, len(Vf)):
  eVf[i] = e*Vf[i]

#encontrando os parâmetros 'a' e 'b'
a,b = np.polyfit(f, eVf, 1)

print("A constante de Planck é", a,"\ne a função trabalho  do catodo W vale", -b)
```

Nota-se pelo gráfico que o módulo do potencial de freamento $V_F$ cresce linearmente com a frequência, o que nos permite calcular a constante de Planck $h$ e o valor da função trabalho do catodo $W$ por regressão linear. Mesmo com apenas 4 pontos foi possível chegar a um valor muito próximo da constante de Planck, o que significa que a aproximação da curva $eV_F \times f$ é ótima.
